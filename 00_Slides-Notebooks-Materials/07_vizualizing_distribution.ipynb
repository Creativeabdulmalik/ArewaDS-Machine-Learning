# Distribution of Data


- Distribution provides insights into the overall shape and characteristics of the data, such as its central tendency (mean, median), variability (range, standard deviation), and the presence of any outliers.


- In simple terms, the distribution of data tells you how the values in your dataset are arranged. It shows you where most of the data points fall, how spread out they are, and whether there are any unusual values (outliers).



### Key Points:

- **Central Tendency:** This refers to the "center" of the data, often measured by the mean (average), median (middle value), or mode (most frequent value).

- **Variability:** This describes how much the data varies or spreads out. Common measures include range (difference between the highest and lowest values), variance, and standard deviation.


- **Shape of Distribution:** This refers to the overall pattern of the data, which can be symmetrical (like a bell curve), skewed (leaning to one side), or uniform (evenly spread out).

- **Outliers:** These are data points that are significantly different from the rest of the data. They can affect the overall analysis and may indicate special cases or errors in data collection.

### When to Use Distribution Analysis:

- **When you want to understand the overall pattern of your data:** Distribution analysis helps you see how values are spread out, which can reveal trends or patterns.


- **When you need to identify outliers:** By examining the distribution, you can spot unusual values that may need further investigation.


- **When comparing different groups:** Distribution analysis allows you to compare how different datasets behave, such as comparing test scores between two classes or sales figures across different regions.





### Examples of Distribution Analysis

- **Customer Age Distribution:** Analyzing the ages of customers who purchased a product can help businesses understand their target audience better.


- **Student Test Scores:** Examining the distribution of test scores in a class can reveal how well students performed and whether there are any outliers who scored significantly higher or lower than the rest.

- **Sales Data:** Analyzing the distribution of sales figures over a period can help identify trends, seasonal patterns, or unusual spikes in sales.

- **Employee Salaries:** Understanding the distribution of salaries within a company can help identify pay equity issues or areas where salary adjustments may be needed.


### **Common Chart Types and Easy Examples:**

The most commonly used plots for distribution are:

- Histogram

- Box Plot (Box-and-Whisker Plot)

- Violin Plot

- Density Plot (Kernel Density Estimate)

- ECDF (Empirical Cumulative Distribution Function)

### Tough Subject

- Right-skewed distribution

- Scenario: A tough subject where most students score low, few excel

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

scores_tough_subject = np.random.beta(a=1, b=4, size=200) * 100  # scale to 0â€“100

plt.figure()
plt.hist(scores_tough_subject, bins=15, edgecolor='black')
plt.title("Student Scores (Right-Skewed): Tough Subject")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```

## Easy Subject

- Left-skewed distribution

- Scenario: An easy subject or generous grading where most students score high

```python
scores_easy_subject = np.random.beta(a=5, b=1.5, size=200) * 100

plt.figure()
plt.hist(scores_easy_subject, bins=15, edgecolor='black')
plt.title("Student Scores (Left-Skewed): Easy Subject")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```

# Bimodal Distribution
- Two peaks in the data
- A subject with two student groups: prepared vs. unprepared

```python

scores_mixed_ability = np.concatenate([
    np.random.normal(loc=50, scale=8, size=100),   # less prepared students
    np.random.normal(loc=85, scale=5, size=100)    # well-prepared students
])
scores_mixed_ability = np.clip(scores_mixed_ability, 0, 100)

plt.figure()
plt.hist(scores_mixed_ability, bins=15, edgecolor='black')
plt.title("Student Scores (Bimodal): Mixed-Ability Class")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```

## Box Plot: A Simple Summary of Data Distribution

Letâ€™s say youâ€™re a school principal and you just got the test scores for 100 students. You want to:


- Quickly summarize the overall performance
  
- See how spread out the scores are
  
- Check for extremely high or low scores
  
- Compare this class to other classes
  
- A histogram could helpâ€”but it:

Depends on bin size

- Canâ€™t show you exact medians or quartiles

- Is harder to compare across multiple groups

âœ… This is where a box plot shines â€” it's a compact, powerful summary!



ðŸ“˜ What Does a Box Plot Show?

A box plot gives you five important numbers:


| Statistic       | Meaning                                                       |
| --------------- | ------------------------------------------------------------- |
| **Min**         | The lowest non-outlier score                                  |
| **Q1**          | The 25th percentile (bottom 25% cutoff)                       |
| **Median (Q2)** | The middle score (50th percentile)                            |
| **Q3**          | The 75th percentile (top 25% cutoff)                          |
| **Max**         | The highest non-outlier score                                 |
| **Outliers**    | Dots beyond whiskers â€“ students who scored unusually high/low |


### Example 1: Box Plot â€” Student Exam Scores

You're a teacher who just finished grading 100 final exam papers. You want to:

- Summarize the class performance

- Find the middle score (median)

- Detect any very high or very low scores (outliers)

- Quickly compare this class to others later

Why Not Histogram?

- A histogram shows how many students got scores in each range, but:

- It's hard to quickly identify the median

- Outliers arenâ€™t always obvious

- Comparing multiple histograms gets messy

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

# Simulate exam scores: most students around 70-85
scores = np.random.normal(loc=75, scale=10, size=100)
scores[:5] = np.random.uniform(30, 45, size=5)   # Add a few low-score outliers
scores = np.clip(scores, 0, 100)

scores
```

```python
plt.figure(figsize=(8, 4))
plt.boxplot(scores, vert=False)
plt.title("Box Plot: Final Exam Scores (100 Students)")
plt.xlabel("Score")
plt.grid(True)
plt.show()
```

Interpretation:

- The box shows most students scored between ~68 and 82.

- The line inside the box (median) is around 75 â€” typical performance.

- The left dots are students who may have failed â€” possible intervention needed.
