# Data Visualization

> As data scientists, one of our most powerful tools is the ability to visualize data. A well-crafted chart or graph can reveal insights that raw numbers alone cannot. 

> As the saying goes, "A picture is worth a thousand words," and in data science, a good visualization can expose trends, outliers, patterns, and relationships that help us understand the story behind the data.

## Types  

we will explore four key categories of data visualization, each serving a different purpose:


- Visualizing Quantities – Comparing magnitudes.

- Visualizing Distributions – Understanding spread and shape.

- Visualizing Proportions – Showing parts of a whole.

- Visualizing Relationships – Examining correlations and interactions.

## Figures and Subplots

Plots in matplotlib reside within a Figure object. You can create a new figure with `plt.figure:`



```python
fig = plt.figure()
```

#### Figure 

- Think of a figure (plt.figure()) as a blank canvas or a blank sheet of paper where you'll draw your plots.

- It's the top-level container that holds everything (all your charts, graphs, etc.).

- You can have multiple figures open at once, each in its own window.


```python
fig = plt.figure()          # Create a blank figure/canvas
ax1 = fig.add_subplot(2, 2, 1)  # Add the first subplot
```

#### Subplots (Individual Plots)
  
- Subplots are individual graphs/charts that you place on your figure.

- The `add_subplot()` method adds a single subplot to your figure.

- The numbers in (2, 2, 1) mean:

  - First number (2): Number of rows in your subplot grid
  
  - Second number (2): Number of columns in your subplot grid

  - Third number (1): Position of this specific subplot in the grid

- plt.figure has a number of options; notably, figsize will guarantee the figure has a certain size and aspect ratio if saved to disk.

- You can’t make a plot with a blank figure. You have to create one or more subplots using add_subplot:





```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

> These plot axis objects have various methods that create different types of plots, and it is preferred to use the axis methods over the top-level plotting functions like `plt.plot`. For example, we could make a line plot with the plot method (see Data visualization after a single plot):

### Example: Create a figure with 3 subplots (arranged vertically)


```python
fig = plt.figure(figsize=(8, 12))  # 8" wide, 12" tall  

#Line Plot 
ax1 = fig.add_subplot(3, 1, 1)  # 3 rows, 1 column, position 1
random_walk = np.random.standard_normal(50).cumsum()
ax1.plot(random_walk, color="black", linestyle="dashed")
ax1.set_title("1. Random Walk Line Plot")
ax1.set_ylabel("Value")
```

```python
# Histogram (Middle) 
fig = plt.figure(figsize=(8, 12))  # 8" wide, 12" tall

ax2 = fig.add_subplot(3, 1, 2)  # Position 2
normal_data = np.random.standard_normal(100)
ax2.hist(normal_data, bins=20, color="black", alpha=0.3)
ax2.set_title("2. Distribution Histogram")
ax2.set_ylabel("Frequency")

```

```python
# Scatter Plot
fig = plt.figure(figsize=(8, 12))  # 8" wide, 12" tall

ax3 = fig.add_subplot(3, 1, 3)  # Position 3
x_values = np.arange(30)
y_values = x_values + 3 * np.random.standard_normal(30)
ax3.scatter(x_values, y_values)
ax3.set_title("3. Noisy Linear Relationship")
ax3.set_xlabel("X Values")
ax3.set_ylabel("Y Values")

plt.tight_layout()  # Prevents label overlapping
plt.show()
```

3- Figures

```python
fig = plt.figure(figsize=(8, 12))  # 8 inches wide, 12 inches tall

ax1 = fig.add_subplot(3, 1, 1)  # (3 rows, 1 column, position 1)
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r-')  # Red line
ax1.set_title("Quadratic Growth")
ax1.set_ylabel("Y Values")

ax2 = fig.add_subplot(3, 1, 2)  # (3 rows, 1 column, position 2)
ax2.bar(['A', 'B', 'C'], [3, 7, 2], color='blue')  # Blue bars
ax2.set_title("Category Comparison")
ax2.set_ylabel("Counts")

ax3 = fig.add_subplot(3, 1, 3)  # (3 rows, 1 column, position 3)
ax3.scatter(np.random.rand(20), np.random.rand(20), color='green')  # Green dots
ax3.set_title("Random Scatter Plot")
ax3.set_xlabel("X Values")
ax3.set_ylabel("Y Values")

# Final adjustments
plt.tight_layout()  # Automatically spaces subplots nicely
plt.show()
```

# Viz Quantities

- Visualizing Quantities – Comparing magnitudes. **Magnitude** refers to the *size* or *absolute value* of a quantitative measurement. In visualization, it represents:
  
  - How large or small a value is


- **When to Use:**

  - Comparing discrete counts (sales, votes, etc.)
  




- **Data Types:**
  
  - X-axis: Categorical (nominal/ordinal)
   
  - Y-axis: Numerical (discrete/continuous)

### Real-World Examples

- Election results by political party

- Monthly revenue across departments

- Population by country


### A. Bar Charts


**Best for:** Comparing 3+ categories  

**Rules:**

- Sort bars by value (descending) unless data is ordinal

- Always start y-axis at 0

- Use consistent colors


```python
import matplotlib.pyplot as plt

products = ['Laptops', 'Phones', 'Tablets']
sales = [1200, 950, 300]
```

```python
plt.bar(products, sales, color='skyblue')
plt.title('Q3 Sales by Product')
plt.xlabel('Product Category')
plt.ylabel('Units Sold (thousands)')
plt.show()
```

### A. Column Charts

- Best for: Time-based comparisons

- Pro Tip: Use horizontal bars for long category names







```python
months = ['January', 'February', 'March']
revenue = [50000, 75000, 60000]

plt.barh(months, revenue, color='salmon')
plt.title('Q1 Revenue')
plt.xlabel('Revenue ($)')
plt.show()
```

```python
### Dot Plots

- Best for: Precise value comparison with few data points

- Demo Code:
```

```python
countries = ['USA', 'Canada', 'Mexico']
gdp = [21, 1.6, 1.2]  # in trillions

plt.plot(gdp, countries, 'o', markersize=10)
plt.title('GDP Comparison')
plt.xlabel('GDP ($T)')
plt.grid(axis='x', linestyle='--')
plt.show()
```
